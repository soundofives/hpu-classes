# Calculation 1

```python
N = 50
u = np.random.rand(N)
u[0] = 0.0
u[N-1] = 1.0

plt.plot(u)
```

This is NOT a straight line! We want to use the method of relaxation to take any configuration and forcing it into a configuration which solves the Laplace equation.

```python
unew = np.zeros(N)
unew[1:N-1] = 0.5*(u[2:N]+u[0:N-2])
unew[0] = 0.0
unew[N-1] = 1.0
plt.plot(unew)
```

This is better--but still not ideal.

```python
i = 0
iteration = 2000
uloop = u
while i < iteration:
    uloop[1:N-1] = 0.5*(uloop[2:N]+uloop[0:N-2])
    uloop[0] = 0.0
    uloop[N-1] = 1.0
    i = i+1
plt.plot(uloop)
```

Around 2000 iterations brings this to a straight line, though any large number would also work! I'm assessing good convergence very loosely here--just as a visually straight line. 

# Calculation 2

```python
N = 100

u = np.random.rand(N,N)
u[0,:] = 0.0
u[:,0] = 0.0
u[N-1,:] = 0.0
u[:,N-1] = 0.0
```

```python
unew = np.zeros((N,N))
unew[1:N-1,1:N-1] = 0.25*(u[2:N,1:N-1]+u[0:N-2,1:N-1]+u[1:N-1,2:N]+u[1:N-1,0:N-2])+0.25*c*rho[1:N-1,1:N-1]
u[0,:] = 0.0
u[:,0] = 0.0
u[N-1,:] = 1.0
u[:,N-1] = 1.0
u = unew

plt.imshow(u)
```

There definitely needs to be more iterations for convergence. Right now, the form is still very random. More iterations would smooth out the plot to a uniform solution.

```python
rho = np.zeros((N,N))
rho[47,:] = 1.0
rho[53,:] = -1.0
c = 1.0
```

c is a multiplying constant with units of meters / Farad, so that when multiplied by a rho it can add to a value with units of Volts.This can represent some initial dampening or amplifying of charge induced in the system.

```python
i = 0
iteration = 2000
uloop = u
while i < iteration:
    uloop[1:N-1,1:N-1] = 0.25*(u[2:N,1:N-1]+u[0:N-2,1:N-1]+u[1:N-1,2:N]+u[1:N-1,0:N-2])+0.25*c*rho[1:N-1,1:N-1]
    uloop[0,:] = 0.0
    uloop[:,0] = 0.0
    uloop[N-1,:] = 1.0
    uloop[:,N-1] = 1.0
    u = uloop
    i = i+1
    
plt.imshow(uloop)
```

The general idea matches exactly what I would expect, but what is the exact nature of the electric field between the plates? I've created a plot below to explore this.

```python
plt.plot(range(0,N),u[47,:])
plt.plot(range(0,N),u[48,:])
plt.plot(range(0,N),u[49,:])
plt.plot(range(0,N),u[50,:])
plt.plot(range(0,N),u[51,:])
plt.plot(range(0,N),u[52,:])
plt.plot(range(0,N),u[53,:])
```

This shows how the E field changes between the plates!

# Calculation 3

```python
xmax = 15
Uvals = np.zeros(xmax)
for x in range(1,xmax):
    N = 200

    u = np.random.rand(N,N)
    u[0,:] = 0.0
    u[:,0] = 0.0
    u[N-1,:] = 0.0
    u[:,N-1] = 0.0

    py = N//2
    px1 = N//2-x
    px2 = N//2+x

    c = 1.0
    rho = np.zeros((N,N))
    rho[int(py),int(px1)] = c
    rho[int(py),int(px2)] = -c

    i = 0
    iteration = 20000
    #uloop = u
    uloop = np.zeros((N,N))
    while i < iteration:
        uloop[1:N-1,1:N-1] = 0.25*(u[2:N,1:N-1]+u[0:N-2,1:N-1]+u[1:N-1,2:N]+u[1:N-1,0:N-2])+(0.25*c*rho[1:N-1,1:N-1])
        uloop[0,:] = 0.0
        uloop[:,0] = 0.0
        uloop[N-1,:] = 0.0
        uloop[:,N-1] = 0.0
        u = uloop
        i = i+1
    #plt.figure()
    #plt.imshow(uloop)
    Ex = 0.5*(u[1:N-1,2:N]-u[1:N-1,0:N-2])
    Ey = 0.5*(u[2:N,1:N-1]-u[0:N-2,1:N-1])
    U = 0.5*np.sum(np.sum(Ex*Ex+Ey*Ey))
    Uvals[x] = U
print(Uvals)
```
