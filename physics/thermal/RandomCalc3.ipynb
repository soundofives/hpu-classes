```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math as math
```

```python
##k = 8.617e-5 ##ev/K
k = 1.38e-23 ##J/K
T = 300 ##K
vp = 0

mN2 = (14.00*2)*(10**-3)/(6.02*10**23)
```

```python
def P(m,k,T,v,vfin,dv):
    P = 0
    while(v < vfin):
        #print(P)
        P = P+((m/(2*np.pi*k*T))**(3/2))*4*np.pi*(v**2)*np.exp((-m*v**2)/(2*k*T))*dv
        v=v+dv
        ##Dv = ((m/(2*np.pi*k*T))**(3/2))*4*np.pi*(v**2)*np.exp**((-m*v**2)/(2*k*T))
        ##print(P)
    
    return P
```

```python
P(mN2, k, T, 399, 401, 0.001)
```

```python
##Faster than 1000 mps

1-P(mN2, k, T, 0, 1000, 0.1)
```

```python
P(mN2, k, T, 12, 47, 0.1)
##probability that it moves between 12 mps and 47 mps
```

```python
def Dv(m,k,T,v):
    Dv = ((m/(2*np.pi*k*T))**(3/2))*4*np.pi*(v**2)*np.exp((-m*v**2)/(2*k*T))
    return Dv
```

```python
probs = []

for i in range(0, 1500):
    probs.append(Dv(mN2,k,T,i))
        
yval = probs
xval = range(0,1500)

fig = plt.figure(figsize=(10,10))
plt.plot(xval, yval)
```

```python
##6.38
##Fraction of molecules moving less than 300 mps
P(mN2, k, T, 0, 300, 0.01)
```

```python
probs2 = []

for i in range(0, 1500):
    probs2.append(Dv(mN2,k,T,i))

probs3 = []

for i in range(0,1500):
    probs3.append(Dv(mN2, k, 600, i))
    
yval2 = probs2
yval3 = probs3
xval = range(0,1500)

fig = plt.figure(figsize=(10,10))
plt.plot(xval, yval2)
plt.plot(xval, yval3)
```

```python
##An increased temperature shifts the distribution to the right, evening it towards the center of the range.
##As more energy is added to the system, the number of microstates and states in general available to the system
##increases, so the distribution in the velocity range is stretched.
```

```python
probs4 = []

for i in range(0, 1500):
    probs4.append(Dv(mN2,k,T,i))

probs5 = []

for i in range(0,1500):
    probs5.append(Dv(mN2*5, k, T, i))
    
yval4 = probs4
yval5 = probs5
xval = range(0,1500)

fig = plt.figure(figsize=(10,10))
plt.plot(xval, yval4)
plt.plot(xval, yval5)
```

```python
##As mass increases, the number of ways to distribute energy in the system increases as well, compressing the number of ways
##to arrange that energy. This reduces the number of states available, making some velocities much more prevelant than others.
```
