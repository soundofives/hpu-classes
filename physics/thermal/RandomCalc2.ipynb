```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math as math
get_ipython().magic('matplotlib inline')
```

```python
#Macrostate calculation
def macrostates(qt):
    return qt+1
```

```python
#Omega
def microstates(n, q):
    omega = math.factorial(n+q-1)/(math.factorial(q)*math.factorial(n-1))
    return omega
```

```python
#Entropy
k = 1.38e-23
def entropy(n,q):
    Sk = math.log(microstates(n,q)) #S in units of S/k
    return Sk
```

```python
#Temp calculation
def temp(N,q):
    
    if(q!=0):
        T = (2)/(entropy(N,q+1)-entropy(N,q-1))
    else:
        T = 0
        
    return T
```

```python
#Cv Calculation
def cv(N,q):
    Cv = 2/(temp(N,q+1)-temp(N,q-1))
    
    return Cv
```
