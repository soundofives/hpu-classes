```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math as math
%matplotlib inline
```

```python
def macrostates(qt):
    return qt+1
```

```python
def microstates(n, q):
    omega = math.factorial(n)/(math.factorial(q)*math.factorial(n-q))
    return omega
```

```python
na = 100
nb = 100
nt = na+nb
qt = 80
qa = 0
qb = 0
```
