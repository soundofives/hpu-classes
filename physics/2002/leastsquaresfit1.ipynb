```python
xdata2 = np.array([399.75,989.25,1578.75,2168.25,2757.75,3346.25,3936.75,4526.25,5115.75,5705.25])
ydata2 = np.array([109,62,39,13,10,4,2,0,1,2])
```

```python
def exponential_fit(x,a,b,c):
    return a*np.exp(-b*x)+c
```

```python
bestparams, covar = curve_fit(exponential_fit, xdata2, ydata2, p0=(1,1e-6,1))
```

```python
bestparams
```

```python
xdata2fit = np.linspace(0,6000,6000)
ydata2fit = exponential_fit(xdata2fit, bestparams[0], bestparams[1], bestparams[2])
```

```python
plt.plot(xdata2, ydata2, 'y*', label='Data Set BEST')
plt.plot(xdata2fit, ydata2fit, 'k', label='Data Set 4')
plt.xlabel('Power Outlets within 10 feet')
plt.ylabel('The Fortitude, Resilience, and Perfection of the Human Form')
plt.title('The Fortitude, Resilience, and Perfection of the Human Form vs Power Outlets within 10 feet')
plt.legend(loc='best', numpoints = 1)
```
