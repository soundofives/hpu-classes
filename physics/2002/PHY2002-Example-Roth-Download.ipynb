```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import *
```

```python
% matplotlib inline
```

```python
x = np.linspace(0,70,500)
y = sin(x)
x2 = np.linspace(0,50,500)
y2 = cos(x2)
```

```python
plt.plot(x,y, 'ks', label='Data Set 1')
plt.plot(x2,y2,'gd', label='Data Set 2')
plt.legend(loc='best', numpoints=1)
plt.xlabel('Second')
plt.ylabel('Lieutenants')
plt.title('Lieutenants per Second')
```
