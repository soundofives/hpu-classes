```python
y = [0, 50, 500, 3400, 5000, 7000, 9000, 400]
x = [0, 1, 20, 25, 40, 47, 49, 50]
y2 = [0, 60, 700, 5000, 2000, 3500, 9500, 9000]
x2 = [0, 1, 20, 25, 40, 47, 49, 50]
plt.plot(x,y, 'y--*', label='Zealots')
plt.plot(x2, y2, 'g--D', label ='Romans')
plt.xlabel('Year')
plt.ylabel('Zealots/Romans')
plt.title('Zealots/Romans per Year')
plt.xlim(0, 55)
plt.ylim(0, 10000)
plt.legend(loc='lower right',numpoints=1)
```

```python
methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']

np.random.seed(47)
grid = np.random.rand(5, 5)

fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})

fig.subplots_adjust(hspace=0.3, wspace=0.05)

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(grid, interpolation=interp_method, cmap='viridis')
    ax.set_title(interp_method)

plt.show()
```
