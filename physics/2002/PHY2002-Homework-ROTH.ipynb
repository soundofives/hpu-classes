```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

from scipy.optimize import curve_fit

% matplotlib inline
```

```python
year = np.array([1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000])
popul = np.array([75.995,91.972,105.711,123.303,131.669,150.697,179.323,203.212,226.505,249.633,281.422])
```

```python
linearfit = np.polyfit(year, popul, 1)
```

```python
linearfit


```

```python
fityear = np.linspace(1900,2000,10000)
fitpop = linearfit[0] * fityear + linearfit[1]
```

```python
plt.scatter(year, popul,label="Population Points")
plt.plot(fityear, fitpop, label="Fit Line")
plt.xlabel("Year")
plt.ylabel("Population of the United States (In Millions)")
plt.title("Population of the United States v. Year")
plt.legend(loc = "best")
```
