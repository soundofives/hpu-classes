```python
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 1
m = 0.001
c = 1
r = 3

dt = 0.1
Nt = 100000
t = np.linspace(0,Nt*dt,Nt)

x1 = np.zeros(Nt)   #position array for left particle
x2 = np.zeros(Nt)   #position array for right particle

v1 = np.zeros(Nt)
v2 = np.zeros(Nt)

y = np.zeros(Nt)    #position array for mediating particle
v = np.zeros(Nt)
```

```python
x1[0] = -r/2
x2[0] = r/2

y[0] = -r/2
v[0] = c
```

```python
i = 0
count =0
while (i<Nt-1):
    
    x1[i+1]=x1[i]+v1[i]*dt
    x2[i+1]=x2[i]+v2[i]*dt
    y[i+1]=y[i]+v[i]*dt
    
    #test for collision between mediating particle and right-moving particle
    if(y[i+1]> x2[i+1]):
        
        v[i+1]= -v[i]
        v2[i+1] = v2[i]+(2*m*c)/M
        v1[i+1] = v1[i]
        count = count+1
          
    #test for collision between mediating particle and left-moving particle
    elif(y[i+1]< x1[i+1]):
        v[i+1]= -v[i]
        v1[i+1] = v1[i]-(2*m*c)/M
        v2[i+1] = v2[i]
        count = count+1
    #right-moving particle maintains same speed
    else:
        v[i+1]=v[i]
        v1[i+1]=v1[i]
        v2[i+1]=v2[i]

    i = i+1    
    #if collision does not occur, all three particles maintain their speeds
print(count)
```

```python
plt.figure(figsize=[20,10])
#plt.xlim(0,200)
plt.plot(t,v2)
plt.title('v vs. t')
plt.xlabel('t')
plt.ylabel('v (m/s)')
plt.show()
```
