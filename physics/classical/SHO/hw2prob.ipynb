```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
t = np.linspace(0,10)

m = .0001 #kg
eta = 1.225*1.8e-5 #kg/m^3  * kg/m.s
R = 0.1 #m
#b = 6*np.pi*eta*R
b = 1.6e-4 #from book for spherical projectile in STP
tau = m/b

g = -9.8 #m/s

vt = tau*g #m/s
v0 = 2*vt
vx0 = 0

xt = lambda x0,vx0,tau,t: vx0*tau*(1-np.exp(-t/tau))
yt = lambda v0,vt,tau,t: (v0+vt)*(1-np.exp(-t/tau))*tau-(vt*t)
vyt = lambda v0,vt,tau,t : (vt)+(v0-vt)*np.exp(-t/tau)

#print(yt(v0,vt,tau,t[10]))

plt.figure(figsize = [20,10])
plt.plot(t,vyt(v0,vt,tau,t))
```

```python
t = np.linspace(0,3)
print(xt(1,1,1,t[1]))
plt.figure(figsize=[20,10])
plt.plot(xt(1,0.3,0.3,t),yt(1,0.3,0.3,t),label='Tau=0.3')
plt.plot(xt(1,1,1,t),yt(1,1,1,t),label='Tau=1')
plt.plot(xt(1,3,3,t),yt(1,3,3,t),label='Tau=3')

plt.ylim(0)
```
